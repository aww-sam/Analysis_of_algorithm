```python
def MaxHeapify(A,n,i):
    largest=i
    l=2*i
    r=(2*i)+1
    
    if (l<=n and A[l]>A[largest]):
        largest=l
    if (r<=n and A[r]>A[largest]):
        largest=r
    
    if largest!=i:
        A[i], A[largest] = A[largest], A[i]
        MaxHeapify(A,n,largest)

def HeapSort(A,n):
    for i in range(n//2,0,-1):
        MaxHeapify(A,n,i)
    for i in range(n,0,-1):
        A[1],A[i]=A[i],A[1]
        MaxHeapify(A,i-1,1)
    return A[1:]    
A = [None,4, 10, 3, 5, 1]
sorted_A = HeapSort(A, len(A) - 1)
print(sorted_A)
```
